The Basic of DiD

In Difference-in-Difference Analysis, we often deal with two main groups: the treatment group and the control group. The treatment group experiences a policy or treatment of interest, while the control group does not. We collect data before and after policy implementation to observe how things change.

image.png

Before you know about Difference-in-Defference...
*   Suppose we wish to estimate the impact of some treatment on a given outcome over time. For example, what is the impact of turtoring on student's GPA?
*   Can we simply calculate the difference in averages between the two groups?
image.png
*  **No**, we need to account for possible different starting point: maybe those in the treatment group already have better grades?
*  DiD will identify the treatment effect as the **difference in changes** over time across groups.




The Parallel Trends Assumption
*   We must assume that the trends of dependent variable over time were identical between the treated and non-treated group prior to treatment.
*   In addition, we assume that the trends would have remained parallel, had it not been for the treatment.
*   Thus, any difference in trends can be attributed to the treatment effect.

The Counterfactual Treatment Group
*   What would the treatment group outcome be post-treatment, if treatment had not occurred?
*   It would, by assumption, follow the same trend as the control group.

Advantages and Limitations of DiD



*   Advantages:
1.   Controls for time trends: DiD inherently controls for factors that change over time but affect both the treatment and control groups equally. These could be things like economic trends or seasonal effects.
2.   Controls for time-invariant group differences: DiD also controls for factors that differ between the treatment and control groups but do not change over time. These could be things like geographic characteristics or cultural differences.
*   Limitations:
1.   Parallel trends assumption: As noted, DiD relies heavily on the assumption that, without the treatment, the treatment and control groups would have followed the same trend over time. If this assumption is violated, the DiD estimate can be biased.
2.   Other time-varying factors: DiD cannot account for other factors that change over time and differently affect the treatment and control groups.
