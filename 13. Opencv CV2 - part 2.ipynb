## 1. Resize
Purpose : Changing the dimensions of an image for better performacne of operations and save the resource
``` python

```

### Functions

There are 2 ways to do image resize `imutils.resize` and `cv2.resize`
- `imutils.resize` evaluate either the width OR height argument, and tries to **preserve the original aspect ratio** of the image
``` python 
imutils.resize(image, width=w/height=h)
```

- `cv2.resize` gives you a free choice (w/o any constraints)
``` python 
cv2.resize(image, (h, w), interpolation=cv2.INTER_CUBIC)
```

| Interpolation  | Note                                                                         |
|:----------------|:-------------------------------------------------------|
| INTER_NEAREST  | a nearest-neighbor interpolation                                             |
| INTER_LINEAR   | a bilinear interpolation (used by default)                                   |
| INTER_AREA     | resampling using pixel area relation. It may be a preferred method for image decimation, as it gives moire’-free results. But when the image is zoomed, it is similar to the INTER_NEAREST method. |
| INTER_CUBIC    | a bicubic interpolation over 4×4 pixel neighborhood                           |
| INTER_LANCZOS4 | a Lanczos interpolation over 8×8 pixel neighborhood                           |

## 4. **Morphological Operations** 
Purpose : Used to detach two connected objects and noise removal. 
``` python 
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(7, 7))

img_erode = cv2.erode(img_gray, kernel) 
img_dilate = cv2.dilate(img_gray, kernel)
img_bilatera = cv2.absdiff(img_erode, img_dilate)
```

Function:
Morphological transformations are some operations based on the image shape. Normally performed on `binary images`. It needs two operation:
1. **Structuring Element (kernel)** : Manually created a structuring elements with help of Numpy. It is `NxN rectangular` shape.
``` python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(n, n))
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(n, n))
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(n, n))
```

2. **Operations**
    - `cv2.erode(image, kernel, iterations)` : It erodes away the boundaries of foreground object (Always try to keep foreground in white). Pixels near boundary will be discarded depending upon the size of kernel.
<img src="./res/erode.png" alt="drawing" width= "1000"/><br><br>

    - `cv2.dilate(image, kernel, iterations)`: It increases the white region in the image or size of foreground object increases.
<img src="./res/dilate.png" alt="drawing" width= "1000"/><br><br>

- `cv2.absdiff(img_erode, img_dilate)` :Calculates the per-element absolute difference between two arrays or between an array and a scalar
<img src="./res/absdiff.png" alt="drawing" width= "1000"/>

## 6. HoughLinesP
Purpose : Find staright lines in a binary image by connecting neighbor pixels
``` python
line_HoughLinesP = cv2.HoughLinesP(img_thresh, rho=1, theta=math.pi/180.0, threshold=100, int(1/7*min(h,w)), 3)
```
Function:
``` python
lines = cv2.HoughLinesP( src, rho, theta, threshold, minLineLength , maxLineGap)
```
| Parameter 	| Value 	| Note 	|
|:---------------|:-------	|:--------------------------------------------	|
| src 	|  	| 8-bit, single-channel binary source image 	|
| lines 	|  	| Output vector of lines. Each line is represented by a 4-element vector (x1,y1,x2,y2) where(x1,y1) and (x2,y2) are the ending points of each detected line segment. 	|
| rho ($\rho$)	|  	| Distance resolution of the accumulator in pixels 	|
| theta ($\theta$)	|  	| Angle resolution of the accumulator in radians 	|
| threshold |  	| Accumulator threshold parameter. Only those lines are returned that get enough votes (>Threshold) 	|
| minLineLength 	| int, 0<x<min(h,w)	| Minimum line length. Line segments shorter than that are rejected. 	|
| maxLineGap | int, 0<x<min(h,w) | Maximum allowed gap between points on the same line to link them 	|


#### Equation of a line in polar coordinates :
<br/>

<br/>

# Appendix3 : Text Detection

````python
# https://github.com/ZER-0-NE/EAST-Detector-for-text-detection-using-OpenCV

import numpy as np
import cv2
from imutils.object_detection import non_max_suppression    # pip install imutils

path = "./res/silicon_valley_2.jpg"
image = cv2.imread(path)
# image1 = cv2.imread("../dedat/431_530_done/1421726.jpg")   # image1是源图像
(height1, width1) = image.shape[:2]
size = 320
(height2, width2) = (size, size)  # 模型输入图像的宽度和高度必须是32的倍数，否则会报错
image2 = cv2.resize(image, (width2, height2))  # image2是现在的图像

net = cv2.dnn.readNet("./res/frozen_east_text_detection.pb")
blob = cv2.dnn.blobFromImage(image2, 1.0, (width2, height2), (123.68, 116.78, 103.94), swapRB=True, crop=False)
net.setInput(blob)

(scores, geometry) = net.forward(["feature_fusion/Conv_7/Sigmoid", "feature_fusion/concat_3"])
(rows, cols) = scores.shape[2:4]  # grab the rows and columns from score volume
rects = []  # stores the bounding box coordiantes for text regions
confidences = []  # stores the probability associated with each bounding box region in rects

for y in range(rows):
    scoresdata = scores[0, 0, y]
    xdata0 = geometry[0, 0, y]
    xdata1 = geometry[0, 1, y]
    xdata2 = geometry[0, 2, y]
    xdata3 = geometry[0, 3, y]
    angles = geometry[0, 4, y]

    for x in range(cols):

        if scoresdata[x] < 0.95:  # if score is less than min_confidence, ignore
            continue
        # print(scoresdata[x])
        offsetx = x * 4.0
        offsety = y * 4.0
        # EAST detector automatically reduces volume size as it passes through the network
        # extracting the rotation angle for the prediction and computing their sine and cos

        angle = angles[x]
        cos = np.cos(angle)
        sin = np.sin(angle)

        h = xdata0[x] + xdata2[x]
        w = xdata1[x] + xdata3[x]
        #  print(offsetx,offsety,xdata1[x],xdata2[x],cos)
        endx = int(offsetx + (cos * xdata1[x]) + (sin * xdata2[x]))
        endy = int(offsety + (sin * xdata1[x]) + (cos * xdata2[x]))
        startx = int(endx - w)
        starty = int(endy - h)

        # appending the confidence score and probabilities to list
        rects.append((startx, starty, endx, endy))
        confidences.append(scoresdata[x])

# applying non-maxima suppression to supppress weak and overlapping bounding boxes
boxes = non_max_suppression(np.array(rects), probs=confidences)

text_detection = image.copy()
rW = width1 / float(width2)
rH = height1 / float(height2)
for (startx, starty, endx, endy) in boxes:

    startx = int(startx * rW)
    starty = int(starty * rH)
    endx = int(endx * rW)
    endy = int(endy * rH)
#     print("startx, starty = ", startx, starty)
    cv2.rectangle(text_detection, (startx, starty), (endx, endy), (0, 255, 255), 15)

plt.imshow(cv2.cvtColor(text_detection, cv2.COLOR_BGR2RGB))
````
